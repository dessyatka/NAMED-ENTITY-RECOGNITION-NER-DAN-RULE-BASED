```python
import pandas as pd
df_latlong=pd.read_csv('data/Indonesian_Family_Life_Survey_4_Longitude_and_Latitude.csv')
df_latlong
```

```python
array_district_temp=df_latlong['kabupaten_name'].str.lower().to_numpy()
array_latitude_temp=df_latlong['lat'].to_numpy()
array_longitude_temp=df_latlong['lng'].to_numpy()

array_district=[]
array_latitude=[]
array_longitude=[]
for i in range (len(array_district_temp)):
    if (array_district_temp[i] not in array_district):
        array_district.append(array_district_temp[i].replace('kota ', '').replace('kabupaten ', ''))
        array_latitude.append(array_latitude_temp[i])
        array_longitude.append(array_longitude_temp[i])
            
        
```

```python
d  = {'district': array_district, 'latitude': array_latitude, 'longitude': array_longitude,}

df_location = pd.DataFrame()
df_location = pd.DataFrame(d, columns=['district','latitude', 'longitude'])
df_location.to_csv('output/df_location2.csv')
df_location
```

```python
df_1  = pd.read_csv('output/df_location1.csv')
df_2 = pd.read_csv('output/df_location2.csv')
```

```python
df_1['count'] = df_1.groupby(['district','latitude', 'longitude']).cumcount()
df_2['count'] = df_2.groupby(['district','latitude', 'longitude']).cumcount()

df_location = pd.concat([df_1,df_2], ignore_index=False)
df_location = df_location.drop_duplicates()
df_location = df_location.drop(['count'], axis=1)
```
